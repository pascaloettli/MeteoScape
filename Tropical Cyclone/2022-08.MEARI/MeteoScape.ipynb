# MeteoScape of TC Meari

```python
import sys
sys.path.insert(0,"../.")
import meteoscape
import anndata
import pandas as pd
import numpy as np
import scvelo as scv
import h5py
import hdf5plugin
```

```python
adata_std = anndata.read_h5ad("adata.h5ad")
basis = 'std'
HD_rate = 0.0
```

# Input

```python
cluster_key = 'clusters'
scv.pl.velocity_embedding_stream(
    adata_std, 
    basis=basis, 
    size=100, 
    title="", 
    color=cluster_key, 
    fontsize=18,
    legend_fontsize=18,
    show=True,
    figsize=(10, 16),
    add_margin=.05
)
```

# Hodge Decomposition

## knn and 15 neighbors

```python
meteoscape.Hodge_decomposition(
    adata_std, 
    basis=basis, 
    HD_rate=HD_rate, 
    graph_method="knn", 
    n_neighbors=15
)
```

### Save

```python
adata_std.write_h5ad("adata.KNN.HD.h5ad")
```
